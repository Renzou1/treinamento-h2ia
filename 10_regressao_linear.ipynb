# Regressao linear no database sklearn.diabetes

```python
from sklearn import datasets
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
diabetes = datasets.load_diabetes()
inputs = diabetes.data
target = diabetes.target
```

```python
ones_column = np.ones(len(inputs))
ones_column = ones_column[:, np.newaxis]
matrix = np.append(ones_column, inputs, 1) # inserts constants
transposed = np.transpose(matrix)

# formula
b = np.matmul(np.linalg.inv(np.matmul(transposed, matrix)), np.dot(transposed, target))
```

```python
def get_prediction(inputs, b):
    modified = np.append(1, inputs) # multiplies first of b by constant
    return np.sum(modified * b)
```

```python
outputs = []
error = []

for index, input in enumerate(inputs):
    prediction = get_prediction(input, b)
    outputs.append(prediction)
    error.append(target)
```

```python
plt.scatter(range(len(target)), target)
plt.scatter(range(len(target)), outputs)
```

```python
plt.scatter(range(len(error)), error)
```
